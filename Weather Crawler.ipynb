```python
main_arr = []
for i in range(1,31):
    current_date = '201904' + str(i).zfill(2)
    #print(current_date)
    url = 'https://www.timeanddate.com/weather/singapore/singapore/historic?month=4&year=2019&hd=' + current_date
    print(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    
    table = soup.find( "table", {"id":"wt-his"} )
    rows=list()

    hourly_time = ['1:00',
                   '2:00',
                   '3:00',
                   '4:00',
                   '5:00',
                   '6:00',
                   '7:00',
                   '8:00',
                   '9:00',
                   '10:00',
                   '11:00',
                   '12:00',
                   '13:00',
                   '14:00',
                   '15:00',
                   '16:00',
                   '17:00',
                   '18:00',
                   '19:00',
                   '20:00',
                   '21:00',
                   '22:00',
                   '23:00']

    # To skip headers
    header_counter = 2
    counter = 0
    for row in table.findAll("tr"):
        #print(row)
        try:
            if counter > header_counter:
                time = row.findAll('th')[0]
                time = str(time)
                time = time.replace('<th>', '')
                time = time.replace('.</th>', '')
                #print(time)
                if time in hourly_time:
                    #print(row)
                    temp_arr = []
                    weather_string = row.findAll('td')[2]
                    weather_string = str(weather_string).replace('<td class="small">','')
                    weather_string = weather_string.replace('</td>','')
                    #print(time,weather_string)
                    temp_arr.append(current_date)
                    temp_arr.append(time)
                    temp_arr.append(weather_string)
                    main_arr.append(temp_arr)
                    #print('')
        except:
            pass
        counter += 1
```
